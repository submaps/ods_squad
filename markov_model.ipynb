```python
import pandas as pd
import numpy as np
import markovify
from tqdm import tqdm_notebook as tqdm
```

```python
%%time
ifile_path = 'data/ods_slack_all2.csv'
df = pd.read_csv(ifile_path, dtype={'ts': str})
df['user_ts'] = df.apply(lambda x: f"{x['user']}_{x['ts']}", axis=1)
df.index = df['user_ts']
```

```python
df_reply = df.query('reply_count > 0').query('channel == "nlp"')
print('df reply count:', len(df_reply))
print(df_reply['text'].head())
```

```python
for msg in df_reply.iterrows():
    print(msg)
    break
```

```python
df_reply
```

```python
%%time
def get_msg_ind(repl_msg_dict):
    return f"{repl_msg_dict['user']}_{repl_msg_dict['ts']}"

lines = []
for ind, msg in tqdm(df_reply.iterrows()):
    joined_reply_text = ''
    for repl_msg_dict in eval(msg['replies']):
        repl_msg_ind = get_msg_ind(repl_msg_dict)
        repl_msg = df.loc[repl_msg_ind]
        joined_reply_text += f". {repl_msg['text']}"
    lines.append({'question': msg['text'], 'answer': joined_reply_text})

df_question_answer_text = pd.DataFrame(lines)
```

```python
df_question_answer_text
```

```python
df_question_answer_text['total_text'] = df_question_answer_text.apply(lambda x: f"{x['question']}. {x['answer']}", axis=1)
```

```python
df_question_answer_text
```

```python
all_text = df_question_answer_text['total_text'].str.join(sep='').sum()
```

```python
len(all_text)
```

```python
all_text[:100]
```

```python
text_model = markovify.Text(all_text.lower(), retain_original=False)
```

```python
input_questions = ['Как завести BERT?',
                   'Кто-нибудь использовал ELMO?',
                   'Делаю сентимент новостей о крипте ', 
                   'Привет! Подскажите, какие вы знаете генеративные nlp модели в проде?']
for input_question in input_questions:
    generated_answer = text_model.make_sentence(init_text=input_question.lower(), max_words=256)
    print(input_question)
    print('\t\t', generated_answer, '\n\n')
```
